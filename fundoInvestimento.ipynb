```python
# Importando a biblioteca
import yfinance as yf
```

```python
# Definindo a carteira de ações
carteira_yf = ['ABEV3.SA', 'B3SA3.SA', 'ELET3.SA', 'GGBR4.SA', 'ITSA4.SA',
               'PETR4.SA', 'RENT3.SA', 'SUZB3.SA', 'VALE3.SA', 'WEGE3.SA']
```

```python
# Carregando os dados da carteira
df = yf.download(carteira_yf, start="2022-08-01", end="2023-08-01")
```

```python
df.head(3)
```

```python
# Passando os ativos para o multindex do df
cotacoes = df.stack (level = 1)
cotacoes
```

```python
# Verificando colunas antes de renomear
cotacoes = cotacoes.reset_index()

# Verifique se 'level_1' está nas colunas
if 'level_1' in cotacoes.columns:
    cotacoes = cotacoes.rename(columns={'level_1': 'Ativo'})
elif 'Ativo' not in cotacoes.columns:
    # Adicione manualmente se necessário
    cotacoes['Ativo'] = 'N/A'  # ou insira uma lógica adequada

# Verifique colunas disponíveis
print("Colunas disponíveis:", cotacoes.columns)

# Agora, selecione apenas as colunas que realmente existem
colunas_esperadas = ["Date", "Open", "High", "Low", "Close", "Ativo"]
colunas_disponiveis = [col for col in colunas_esperadas if col in cotacoes.columns]
cotacoes = cotacoes[colunas_disponiveis]

# Exibir resultado
print(cotacoes.head())
```
